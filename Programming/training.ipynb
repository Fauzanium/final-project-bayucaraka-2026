```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO
model = YOLO("best.pt")
img = cv2.imread("./runs/detect/predict/images/IMG_20240917_143644.jpg")
results = model(img)
box = results[0].boxes.xyxy[0].cpu().numpy().astype(int)
x1, y1, x2, y2 = box
u = (x1+x2) / 2
v = (y1+y2) / 2
fx = 1000
fy = 1000
cx = img.shape[1] / 2
cy = img.shape[0] / 2
H = 1.0
x_norm = (u - cx) / fx
y_norm = (v - cy) / fy

X= x_norm * H
Y= y_norm * H
Z= 0

cv2.circle(img, (int(u), int(v)), 5, (0,0,255), -1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title(f"X={X:.2f}, Y={Y:.2f}, Z={Z:.2f}")
plt.axis("off")
```
